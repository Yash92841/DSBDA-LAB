```python
df["Writing_Score"] =df["Writing_Score"].fillna(df["Writing_Score"].std())
```

```python
df['Math_Score'] = df['Math_Score'].astype(int)
df['Reading_Score'] = df['Reading_Score'].astype(int)
df['Writing_Score'] = df['Writing_Score'].astype(int)
```

```python
sorted_score =sorted(df['Reading_Score'])
sorted_score
```

```python
q1 = np.percentile(sorted_score, 25)
q3 = np.percentile(sorted_score, 75)
print(q1,q3)
```

```python
IQR = q3-q1
```

```python
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
r_outliers = []
for i in sorted_score:
    if (i<lwr_bound or i>upr_bound):
        r_outliers.append(i)
print(r_outliers)
```
